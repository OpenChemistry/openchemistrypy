```python
import openchemistry as oc
```

```python
calc_setup = {
    'basis': '3-21g',
    'theory': 'dft'
}
```

```python
mymol = oc.find_structure('RYYVLZVUVIJVGH-UHFFFAOYSA-N')
```

```python
result = mymol.optimize(**calc_setup)
```

```python
result.structure.show()
```

```python
result = mymol.frequencies(**calc_setup)
```

```python
result.frequencies.show()
```
